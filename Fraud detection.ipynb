# 2.Data Preprocessing


```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

data.isnull().sum()

# scale time and amount for better performance
scaler=StandardScaler()
data[['Time','Amount']]=scaler.fit_transform(data[['Time','Amount']])

# handle class imbalance using SMOTE
X=data.drop('Class',axis=1)
y=data['Class']
smote=SMOTE(random_state=42)
X_resampled,y_resampled=smote.fit_resample(X,y)

print("Clas distribution after SMOTE:",y_resampled.value_counts())
```

# 3.Exploratory Data Analysis(EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# plot class distribution
sns.countplot(x=y_resampled)
plt.title("Class Distribution")
plt.show()


# Correlation heatmap

plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),cmap='coolwarm',annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

# 4. Feature Engineering

```python
from sklearn.decomposition import PCA

# # Apply PCA to reduce dimensions if needed
pca=PCA(n_components=10)
X_pca=pca.fit_transform(X_resampled)
print("Explained variance Ratios:",pca.explained_variance_ratio_)
```

# 5.Model Training

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
print("Logistic Regression Report:")
print(classification_report(y_test, y_pred_lr))
```

```python
# Random Forest
rf = RandomForestClassifier()
param_grid = {'n_estimators': [100, 200], 'max_depth': [10, 20]}
grid_rf = GridSearchCV(rf, param_grid, cv=3)
grid_rf.fit(X_train, y_train)
y_pred_rf = grid_rf.best_estimator_.predict(X_test)
```

# 6. Model Evaluation

```python
# Confusion Matrix for Random Forest 
conf_matrix=confusion_matrix(y_test,y_pred_rf)
sns.heatmap(conf_matrix, annot=True, fmt='d',cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```

```python
import pickle

# Save the best estimator from grid search
with open('fraud_detection_model.pkl', 'wb') as f:
    pickle.dump(grid_rf.best_estimator_, f)
```
